# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
#create path to DB
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#get table names in hawaii schema
inspector = inspect(engine)
inspector.get_table_names()
```

```python
#columns in measurement table
columns_meas = inspector.get_columns('measurement')
for c in columns_meas:
    print(c['name'], c['type'])
```

```python
#columns in Station table
columns_station = inspector.get_columns('station')
for c in columns_station:
    print(c['name'], c['type'])
```

```python
# Find Latest Date
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
last_date = last_date[0]

#Deconstrut last date into yr, mo, day
last_yr = int(last_date[0:4])
last_mo = int(last_date[5:7])
last_day = int(last_date[-2:])
```

```python
#Find date that is 365 days prior to last date
query_date = str(dt.date(last_yr, last_mo, last_day) - dt.timedelta(days=366))

# Design a query to retrieve the last 12 months of precipitation data, ordered by date
precip_data = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > query_date).\
    order_by(Measurement.date).all()

#print 5 from the list
precip_data[0:5]
```

```python
#Find date that is 365 days prior to last date
query_date = str(dt.date(last_yr, last_mo, last_day) - dt.timedelta(days=366))

# Design a query to retrieve the last 12 months of precipitation data, ordered by date
precip_data = session.query(Measurement.tobs).\
    filter(Measurement.date > query_date).\
    order_by(Measurement.date).all()

#print 5 from the list
precip_data[0:5]
```

```python
engine.execute('Select * from measurement where date > "2016-08-26" order by date').fetchall()
engine.execute('Select * from measurement order by date').fetchall()
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
precip_df = pd.DataFrame(precip_data, columns = ['date' , 'precipitation'])

precip_df.set_index('date',inplace=True)

precip_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
#plt.bar(precip_df.index, precip_df.precipitation, color='b', alpha=0.5, align="center")
plt.plot(precip_df.index, precip_df.precipitation, color='blue', alpha = 0.5)

plt.grid(True)
plt.title("Date vs Precipitation")
plt.xlabel("Date")
plt.ylabel("Precipitation")
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precip_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
x = session.query(Measurement.station).\
    group_by(Measurement.station).count()
print ('Number of Stations: ', x)
```

```python
# Design a query to show how many stations are available in this dataset?
x = session.query(Station.station).\
    group_by(Station.station).count()
print ('Number of Stations: ', x)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).\
group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
#lowest temp at most active station
low_temp= session.query(func.min(Measurement.tobs)).filter(Measurement.station=='USC00519281').first()
#Highest temp at most active station
high_temp = session.query(func.max(Measurement.tobs)).filter(Measurement.station=='USC00519281').first()
#average temp
avg_temp = session.query(func.avg(Measurement.tobs)).filter(Measurement.station=='USC00519281').first()
print('Lowest Temp: ', low_temp)
print('Highest Temp: ', high_temp)
print('Avg Temp: ', np.round(avg_temp,2))
```

```python
# list stations in order of freq of observations during the last 12months
session.query(Measurement.station, func.count(Measurement.station)).\
    filter(Measurement.date > query_date).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
```

```python
# Query the last 12 months of temperature observation data for station with most observations
# and plot the results as a histogram
temp_data = session.query(Measurement.tobs).filter(Measurement.date > query_date, Measurement.station == 'USC00519281').all()
```

```python
#store data in a dataframe
temp_df = pd.DataFrame(temp_data, columns = ['temp'])

#plot histogram
plt.hist(temp_df['temp'], bins=12)
```
